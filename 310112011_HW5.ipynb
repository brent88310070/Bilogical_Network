```python
import networkx as nx
import scipy as sp
import pandas as pd
import numpy as np
```

### Find  1  isomorphic  graph  of  network  1

```python
G = nx.read_edgelist("Network 1.txt", create_using=nx.MultiGraph())
```

```python
nx.draw(G, with_labels = True)
```

```python
G_iso = nx.read_edgelist("iso_Network 1.txt", create_using=nx.MultiGraph())
```

```python
nx.draw(G_iso, with_labels = True)
```

```python
nx.is_isomorphic(G, G_iso)
```

### List  the  deck  of  network  1

```python
all_nodes = list(G.nodes)
print("Network 1: Decks show by adjacency list format")
for i in range(len(all_nodes)):
    deck = G.copy()
    deck.remove_node(all_nodes[i])
    print("#Deck %s" %i)
    for line in nx.generate_adjlist(deck):
        print(line)
```

### Check  if  the  decks  of  the  isomorphic  graphs  you  found are  identical  to  the  deck  of  network  1

```python
def check_iso(G_iso_deck):
    all_nodes = list(G.nodes)
    for i in range(len(all_nodes)):
        deck = G.copy()
        deck.remove_node(all_nodes[i])
        if nx.is_isomorphic(deck, G_iso_deck):
            return True
    else:
        return False
```

```python
iso_all_nodes = list(G_iso.nodes)
check = len(iso_all_nodes)
for i in range(len(iso_all_nodes)):
    G_iso_deck = G_iso.copy()
    G_iso_deck.remove_node(iso_all_nodes[i])
    if check_iso(G_iso_deck) == False:
        check -= 1
        print("Not isomorphic graphs!")
if check == len(iso_all_nodes):
    print("isomorphic graphs!")
```
