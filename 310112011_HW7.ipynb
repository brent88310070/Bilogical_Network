#### Treat networks 3 as a undirected graph
#### is it  strongly  orientable?

```python
G = nx.read_edgelist("Network 3.txt")
```

```python
k = 0
for i in range(len(list(G.edges))):
    G2 = G.copy()
    G2.remove_edge(list(G.edges)[i][0], list(G.edges)[i][1])
    if (nx.is_connected(G2)):
        k+=1
    else:
        print("Not strongly orientable")
if k == len(list(G.edges)):
    print("Strongly orientable")
```

#### if yes, find the strongly oriented graph

```python
G = nx.read_edgelist("Network 3.txt")
nx.draw(G, with_labels = True)
```

```python
check_strong_con = False
while(check_strong_con != True):
    edge = list(G.edges())
    edge_random_chose = []
    for i in range(len(edge)):
        l = [edge[i][0], edge[i][1]]
        random.shuffle(l)
        edge_random_chose.append(l)
        
    g = nx.DiGraph()
    g.add_nodes_from(list(G))
    g.add_edges_from(edge_random_chose)
    check_strong_con = nx.is_strongly_connected(g)
nx.draw(g, with_labels = True)
```

#### and, find the vertex-cut and its vertex-connectivity

```python
vertex_cut = []
vertex_connectivity = []
for i in list(g.nodes()):
    gcut = g.copy()
    gcut.remove_node(i)
    check_strong_con = nx.is_strongly_connected(gcut)
    if(check_strong_con != True):
        vertex_cut.append(i)

for i in list(g.edges()):
    gcon = g.copy()
    gcon.remove_edge(i[0], i[1])
    check_strong_con = nx.is_strongly_connected(gcon)
    if(check_strong_con != True):
        vertex_connectivity.append(i)
        
print('vertex-cut:',vertex_cut)
print('vertex-connectivity:',vertex_connectivity)
```
